```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
HURDAT2 = "hurdatawdates.csv"
storm_month_df = pd.read_csv(HURDAT2, low_memory=False)

storm_month_df.head()
```

```python
storm_month = storm_month_df['Month'].value_counts()
# storm_month.count().head()
```

```python
# monthly_storms = storm_month['Year'].count()
```

```python
storm_occurance = storm_month.to_frame()

storm_occurance.reset_index(inplace=True)
```

```python
%matplotlib notebook
```

```python
x = storm_occurance["index"]
y = storm_occurance["Month"]
plt.bar(x, y)

plt.title("Average Number of Storm Occurances by Month")
plt.xlabel("Month")
plt.ylabel("Average Number of Storms")
```

```python
output_file = "CSV_Output/Images/Average_Number_of_Storm_Occurances.png"
plt.savefig(output_file)
plt.show()
```
